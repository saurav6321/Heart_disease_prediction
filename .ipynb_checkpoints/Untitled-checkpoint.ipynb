# splitting dataset 

```python

#Create independent and Dependent Features
columns = data.columns.tolist()
# Filter the columns to remove data we do not want 
columns = [c for c in columns if c not in ["TenYearCHD"]]
# Store the variable we are predicting 
target = "TenYearCHD"
# Define a random state 
state = np.random.RandomState(42)
x= data[columns]
y = data[target]
# Print the shapes of X & Y
print(x.shape)
print(y.shape)
```

# filling up null values

```python
x.isnull().sum()
```

```python
y.isnull().sum()
```

```python
x['cigsPerDay'].fillna(x.groupby('currentSmoker')['cigsPerDay'].transform('mean'),inplace=True)
x['BPMeds'].fillna(x.groupby('prevalentHyp')['BPMeds'].transform('median'),inplace=True)
x['totChol'].fillna(x.groupby('prevalentStroke')['totChol'].transform('mean'),inplace=True)
x['BMI'].fillna(x.groupby('age')['BMI'].transform('mean'),inplace=True)
x['heartRate'].fillna(x.groupby('sysBP')['heartRate'].transform('mean'),inplace=True)
x['glucose'].fillna(x.groupby('diabetes')['glucose'].transform('mean'),inplace=True)
x['education'].fillna(x.groupby('age')['education'].transform('mean'),inplace=True)



```

```python
x.isnull().sum()
```

# Balance Dataset

```python
sns.set_style('whitegrid')
sns.countplot(x ='TenYearCHD',data=data,palette='RdBu_r')
```

```python
disease= data[data['TenYearCHD']==1]
no= data[data['TenYearCHD']==0]
print(disease.shape,no.shape)
```

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss
```

```python
smk=SMOTETomek(random_state=42)
x_res,y_res=smk.fit_resample(x,y)
x_res.shape,y_res.shape
```

```python
from collections import Counter
print('original data {}'.format(Counter(y)))
print('Resampled data {}'.format(Counter(y_res)))
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale =['age','BPMeds','cigsPerDay','totChol','sysBP','diaBP','BMI','heartRate','glucose']
x_res[columns_to_scale] = standardScaler.fit_transform(x_res[columns_to_scale])
```

```python
x_res
```

```python
x_res.isnull().sum()
```

# Different Classifiers

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X=x_res.drop(['education','currentSmoker'],axis=1)
Y=y_res
X
```

```python

from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,Y,cv=10)
    knn_scores.append(score.mean())
```

```python

plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = k)
score=cross_val_score(knn_classifier,X,Y,cv=10)
knn_scores.append(score.mean())
knn_scores
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 16)
score=cross_val_score(knn_classifier,X,Y,cv=10)
score.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

randomforest_classifier= RandomForestClassifier(n_estimators=10)

score=cross_val_score(randomforest_classifier,X,Y,cv=10)
```

```python

score.mean()
```

# Train-Test Split


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
x_train
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn import linear_model
reg=linear_model.LogisticRegression()
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
#reg.predict(x_test)
```

```python
clf=RandomForestClassifier(n_estimators = 100) 
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
#clf.predict(x_test)
```

# Confusion matrix

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
clf.fit(x_train,y_train)
predict = clf.predict(x_test)
```

```python
print('\n')
print(classification_report(y_test,predict))
```

```python
def Heart_disease(male,age,cigsPerDay,BPMeds,prevalentStroke,prevalentHyp,diabetes,totChol,sysBP,diaBP,BMI,heartRate,glucose):
    a=np.zeros(len(X.columns))
    a[0]=male
    a[1]=age
    a[2]=cigsPerDay
    a[3]=BPMeds
    a[4]=prevalentStroke
    a[5]=prevalentHyp
    a[6]=diabetes
    a[7]=totChol
    a[8]=sysBP
    a[9]=diaBP
    a[10]=BMI
    a[11]=heartRate
    a[12]=glucose
    return clf.predict([a])
   
```
