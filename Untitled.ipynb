# splitting dataset 

```python

#Create independent and Dependent Features
columns = data.columns.tolist()
# Filter the columns to remove data we do not want 
columns = [c for c in columns if c not in ["TenYearCHD"]]
# Store the variable we are predicting 
target = "TenYearCHD"
# Define a random state 
state = np.random.RandomState(42)
x= data[columns]
y = data[target]
# Print the shapes of X & Y
print(x.shape)
print(y.shape)
```

# filling up null values

```python
x.isnull().sum()
```

```python
y.isnull().sum()
```

```python
x['cigsPerDay'].fillna(x.groupby('currentSmoker')['cigsPerDay'].transform('mean'),inplace=True)
x['BPMeds'].fillna(x.groupby('prevalentHyp')['BPMeds'].transform('median'),inplace=True)
x['totChol'].fillna(x.groupby('prevalentStroke')['totChol'].transform('mean'),inplace=True)
x['BMI'].fillna(x.groupby('age')['BMI'].transform('mean'),inplace=True)
x['heartRate'].fillna(x.groupby('sysBP')['heartRate'].transform('mean'),inplace=True)
x['glucose'].fillna(x.groupby('diabetes')['glucose'].transform('mean'),inplace=True)
x['education'].fillna(x.groupby('age')['education'].transform('mean'),inplace=True)



```

# Different Classifiers

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X=x_res.drop(['education','currentSmoker'],axis=1)
Y=y_res
X=X.astype({"cigsPerDay":'int', "BPMeds":'int',"prevalentStroke":'int',"prevalentHyp":'int',"diabetes":'int',"totChol":'int',"sysBP":'int',"diaBP":'int',"BMI":'int',"heartRate":'int',"glucose":'int'}) 
#Y=Y.astype({"cigsPerDay":'int', "BPMeds":'int',"prevalentStroke":'int',"prevalentHyp":'int',"diabetes":'int',"totChol":'int',"sysBP":'int',"diaBP":'int',"BMI":'int',"heartRate":'int',"glucose":'int'})
X
```

```python

from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,Y,cv=10)
    knn_scores.append(score.mean())
```

```python

plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = k)
score=cross_val_score(knn_classifier,X,Y,cv=10)
knn_scores.append(score.mean())
knn_scores
```

# Train-Test Split


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
x_train
```

```python
clf=RandomForestClassifier(n_estimators = 100) 
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
#clf.predict(x_test)
```
